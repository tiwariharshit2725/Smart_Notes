```python
import numpy as np
import pandas as pd
```

```python
train_df=pd.read_csv("/content/drive/MyDrive/train.csv")
train_df
```

```python
train_df=train_df[:50000]
train_df
```

```python
test_df=pd.read_csv("/content/drive/MyDrive/test.csv")
test_df
```

```python
test_df=test_df[:20000]
```

```python
train_df=train_df.rename(columns={"headline":"target_text","article":"source_text"})
test_df=test_df.rename(columns={"headline":"target_text","article":"source_text"})
train_df=train_df[['source_text','target_text']]
test_df=test_df[['source_text','target_text']]
```

```python
train_df['source_text']="summarize: "+train_df['source_text']
test_df['source_text']="summarize: "+test_df['source_text']
train_df
```

```python
from simplet5 import SimpleT5
```

```python
model=SimpleT5()
model.from_pretrained(model_type="t5",model_name="t5-base")
```

```python
model.train(train_df=train_df[:1000] , eval_df=test_df[:100] , source_max_token_len=128 , target_max_token_len=50 , batch_size=8 , max_epochs=5 , use_gpu=True)
```
